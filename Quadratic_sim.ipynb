```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# choosing a tangent value
y_true = 3.0
```

```python
# creating a range of possible predictions for y_hat (these represent predictions produced by different weights)
y_hat = np.linspace(-1, 7, 400)
y_hat[1:20]
```

```python
# Cost function
# C = 1/2 * (y_hat - y_true)^2
cost = 0.5 * (y_hat - y_true)**2
cost[1:20]
```

```python
# sample data for predictions
sample_preds = np.array([-0.5, 1.0, 2.5, 3.0, 4.5, 6.0])
sample_costs = 0.5 * (sample_preds - y_true)**2
sample_costs
```

```python
# plot
plt.figure(figsize=(8, 4.5))
plt.plot(y_hat, cost, label="Cost curve (0.5*(ŷ - y)^2)")
plt.scatter(sample_preds, sample_costs, s=60, label="Sample prediction points")

plt.title("Quadratic Cost Function Simulation")
plt.xlabel("Predicted value (ŷ)")
plt.ylabel("Cost C")
plt.grid(True)
plt.legend()
plt.show()
```
